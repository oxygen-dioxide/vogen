# PyVogen使用示例：从music21合成歌声

```python
import music21
import vogen
from vogen import synth
#环境配置
#us = music21.environment.UserSettings()
#us['musescoreDirectPNGPath'] = r'E:\Music-----------------\MuseScore\bin\MuseScore3.exe'
#us['musicxmlPath'] =r'E:\Music-----------------\MuseScore\bin\MuseScore3.exe'
```

```python
#乐谱
stream=music21.converter.parse("""tinyNotation: 4/4 r1
G4 E8 G c4. r8
A4 c G4. r8
G4 C8 D E4 D8 C
D2. r4
G4 E8 G c4. B8
A4 c G4. r8
G4 D8 E F4. BB8
C2. r4""", makeNotation=False)
stream.show()
```

```python
#歌词
lyrics="长亭-外，古道边，芳草-碧连-天。晚风-拂柳笛声残，夕阳-山外山。".replace("，","").replace("。","")
#灌入歌词
for n, lyric in zip(stream.notes, lyrics):
    n.lyric = lyric
stream.show()
```

```python
vf=vogen.to_vog_file(stream).lyrictorom()
print(vf)
vf.save("songbie.vog")
synth.play(vf)
```
